```python
# =====================================================================
# STEP 1: IMPORT CORE LIBRARIES AND AUTO-DETECT STUDENT DATASET
# =====================================================================
import pandas as pd
import numpy as np
import os
import glob

# Adjusted folder path to include the exact sub-folder inside dataset
folder_path = r'C:\Users\noorm\Documents\BSIT-F-23-B-937\Semesters\Semester 06\Artificial Intelligence Lab\Final Semester Project\ai-mental-health-companion\backend\dataset\Student Mental health'

# Looking for the .csv file inside that specific folder
matching_files = glob.glob(os.path.join(folder_path, 'Student Mental health*.csv*'))

if not matching_files:
    print("CRITICAL ERROR: No valid .csv file found in the dataset folder. Please verify the folder contents.")
else:
    # Select the first exact matching data file path
    csv_path = matching_files[0]
    try:
        df = pd.read_csv(csv_path)
        print(f"Successfully loaded file from automatic path detection: {csv_path}")
        print("\nDataset Diagnostics:")
        print(df.info())
    except PermissionError:
        print(f"PERMISSION ERROR: Please close '{os.path.basename(csv_path)}' if it is open in Excel or another application, then try again.")
```

```python
# =====================================================================
# STEP 2: EXPLORATORY DATA ANALYSIS (UNIVARIATE & BIVARIATE VISUALS)
# =====================================================================
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Plot 1: Age distribution mapping
plt.figure(figsize=(10, 4))
sns.histplot(df['Age'].dropna(), kde=True, color='blue', bins=10)
plt.title('Age Distribution of Students')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()

# Plot 2: Categorical mapping for mental health columns correlation matrix
mapping_df = df.copy()
mental_health_cols = ['Do you have Depression?', 'Do you have Anxiety?', 'Do you have Panic attack?']

for col in mental_health_cols:
    if col in mapping_df.columns:
        mapping_df[col] = mapping_df[col].map({'Yes': 1, 'No': 0})

plt.figure(figsize=(8, 6))
numerical_df = mapping_df.select_dtypes(include=[np.number])
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Mental Health Features Correlation Matrix')
plt.show()
```

```python
# =====================================================================
# STEP 3: DATA PREPROCESSING, FEATURE SCALING, AND TRAIN-TEST SPLIT
# =====================================================================
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# 1. Fill any missing values in the Age column with the median safely
df['Age'] = df['Age'].fillna(df['Age'].median())

# 2. Map string metrics ('Yes'/'No') to binary targets (1/0)
clean_df = df.copy()
for col in ['Do you have Depression?', 'Do you have Anxiety?', 'Do you have Panic attack?']:
    clean_df[col] = clean_df[col].map({'Yes': 1, 'No': 0})

# 3. Construct a unified risk assessment label based on mental health column states
clean_df['Risk_Bracket'] = np.where((clean_df['Do you have Depression?'] == 1) | (clean_df['Do you have Anxiety?'] == 1), 1, 0)

# 4. Select numerical features matching our specific pipeline variables
features = ['Age', 'Do you have Depression?', 'Do you have Anxiety?', 'Do you have Panic attack?']
X = clean_df[features]
y = clean_df['Risk_Bracket']

# 5. Normalize continuous attributes into a standardized [0, 1] range
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# 6. Partition data into 70% training and 30% testing validation vectors
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)
print(f"Preprocessing Complete. X_train shape: {X_train.shape}, X_test shape: {X_test.shape}")
```

```python
# =====================================================================
# STEP 4: TRAIN RANDOM FOREST CLASSIFIER AND EVALUATE DIAGNOSTICS
# =====================================================================
# Importing all required diagnostic packages locally to eliminate NameError
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize and train the nonlinear ensemble model on training matrices
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Execute risk level predictions across the testing evaluation set
y_pred = rf_model.predict(X_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))

# Map predictions onto a confusion matrix to evaluate diagnostic accuracy
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Low Risk', 'High Risk'], yticklabels=['Low Risk', 'High Risk'])
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
# =====================================================================
# STEP 5: SERIALIZE AND EXPORT MODEL OBJECTS TO THE BACKEND PIPELINE
# =====================================================================
# Import os locally to prevent NameError if cells are run out of order
import os
import pickle

# Establish target architecture paths and freeze model state to disk
os.makedirs('../backend/models', exist_ok=True)
model_export_path = '../backend/models/model.pkl'

with open(model_export_path, 'wb') as f:
    pickle.dump(rf_model, f)

print(f"Model saved successfully to: {model_export_path}")
```
