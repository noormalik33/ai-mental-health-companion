```python
# =====================================================================
# STEP 1: MERGE AND INGEST SEPARATED TEXT DATASTREAMS
# =====================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Your verified absolute path to the dataset folder
folder_path = r'C:\Users\noorm\Documents\BSIT-F-23-B-937\Semesters\Semester 06\Artificial Intelligence Lab\Final Semester Project\ai-mental-health-companion\backend\dataset\Emotions dataset for NLP'

text_files = [
    os.path.join(folder_path, 'train.txt'),
    os.path.join(folder_path, 'val.txt'),
    os.path.join(folder_path, 'test.txt')
]

combined_entries = []

# Aggregate files and extract semantic structures safely
for file_path in text_files:
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f:
                line = line.strip()
                if ';' in line:
                    # rsplit(';', 1) splits from the last semicolon to preserve internal punctuation
                    text_data, emotion_tag = line.rsplit(';', 1)
                    combined_entries.append({
                        'journal_entry': text_data.strip(), 
                        'emotion': emotion_tag.strip()
                    })
        print(f"Successfully loaded file: {os.path.basename(file_path)}")
    except FileNotFoundError:
        print(f"CRITICAL ERROR: Could not find file at: {file_path}")

# Build a single unified qualitative data frame architecture
text_df = pd.DataFrame(combined_entries)

if text_df.empty:
    print("\nERROR: text_df is empty! Please verify that files contain data.")
else:
    print(f"\nTotal rows successfully loaded: {text_df.shape[0]}")
    print("Columns detected:", text_df.columns.tolist())
    print("\nFirst 5 Rows:")
    print(text_df.head())
```

```python
# =====================================================================
# STEP 2: EXPLORATORY DATA ANALYSIS (VISUAL FREQUENCY & WORD CLOUD)
# =====================================================================
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# --- 2.1: EMOTION FREQUENCY DISTRIBUTION COUNTPLOT ---
plt.figure(figsize=(10, 4))
sns.countplot(
    data=text_df, 
    x='emotion', 
    hue='emotion',
    order=text_df['emotion'].value_counts().index, 
    palette='viridis',
    legend=False
)
plt.title('Emotion Labels Frequency Distribution')
plt.xlabel('Emotion Category')
plt.ylabel('Number of Journal Entries')
plt.grid(axis='y', alpha=0.3)
plt.show()

# Print quantitative class metrics for validation
print("Text Class Distribution Metrics:")
print(text_df['emotion'].value_counts())
print("-" * 70)

# --- 2.2: GENERATE TEXT CORPUS WORD CLOUD ---
# Flatten qualitative narratives into a comprehensive text array corpus
full_text_corpus = " ".join(text_df['journal_entry'].tolist())

# Synthesize frequent keywords into an animated graphical structure
wordcloud = WordCloud(
    width=800, 
    height=400, 
    background_color='white', 
    colormap='magma', 
    max_words=100
).generate(full_text_corpus)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Journal Entry Word Cloud')
plt.show()
```

```python
# =====================================================================
# STEP 3: TEXT PREPROCESSING AND FEATURE EXTRACTION VIA TF-IDF
# =====================================================================
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Isolate document elements and target semantic labels
X_text = text_df['journal_entry']
y_text = text_df['emotion']

# Standardize extraction parameters using character bounds and unigram tokens
tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
X_tfidf = tfidf_vectorizer.fit_transform(X_text)

# Partition structured matrices into training vectors (80%) and testing dimensions (20%)
X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(X_tfidf, y_text, test_size=0.2, random_state=42, stratify=y_text)

print(f"Feature Extraction & Train-Test Split Complete.\n")
print(f"X_train_t shape (Training text vectors): {X_train_t.shape}")
print(f"X_test_t shape (Testing text dimensions): {X_test_t.shape}")
```

```python
# =====================================================================
# STEP 4: TRAIN TEXT CLASSIFIER AND EVALUATE NLP DIAGNOSTICS
# =====================================================================
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

print("Training Logistic Regression Text Classifier...\n")

# Initialize and fit the standard classifier model on the training text vectors
text_model = LogisticRegression(max_iter=1000, random_state=42)
text_model.fit(X_train_t, y_train_t)

# Calculate textual category predictions across the test dimensions
y_pred_t = text_model.predict(X_test_t)

print("NLP Text Classification Report:")
print(classification_report(y_test_t, y_pred_t))

# Render classification evaluation metrics inside a confusion heatmap matrix
cm_t = confusion_matrix(y_test_t, y_pred_t)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_t, annot=True, fmt='d', cmap='Purples', cbar=False,
            xticklabels=text_model.classes_, yticklabels=text_model.classes_)
plt.title('Text NLP Confusion Matrix Heatmap')
plt.ylabel('Actual Emotion State')
plt.xlabel('Predicted Emotion State')
plt.show()
```

```python
# =====================================================================
# STEP 5: SERIALIZE AND EXPORT MODEL OBJECTS TO THE BACKEND PIPELINE
# =====================================================================
import os
import pickle

# Establish backend target structure verification paths safely
backend_dir = '../backend'
os.makedirs(backend_dir, exist_ok=True)

model_path = os.path.join(backend_dir, 'emotion_nlp_model.pkl')
vectorizer_path = os.path.join(backend_dir, 'emotion_vectorizer.pkl')

# Serialize trained text classifier pipeline states to disk
with open(model_path, 'wb') as f:
    pickle.dump(text_model, f)

# Serialize operational tfidf vectorizer extraction configurations to disk
with open(vectorizer_path, 'wb') as f:
    pickle.dump(tfidf_vectorizer, f)

print(f"NLP Model saved successfully to target destination: {model_path}")
print(f"TF-IDF Vectorizer saved successfully to target destination: {vectorizer_path}")
```
