```python
# =====================================================================
# STEP 1: IMPORT CORE LIBRARIES AND INGEST STUDENT SURVEY DATASET
# =====================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import pickle
import os

# Define the dynamic file path matching the project directory structure
csv_path = '../Student Mental health.csv'
df = pd.read_csv(csv_path)

# Display dataset diagnostics to inspect rows, columns, and data types
print("Dataset Info:")
print(df.info())
print("\nFirst 5 Rows:")
print(df.head())
```

```python
# =====================================================================
# STEP 2: EXPLORATORY DATA ANALYSIS (UNIVARIATE & BIVARIATE VISUALS)
# =====================================================================
# Generate a histogram to visualize the distribution profile of anxiety scores
plt.figure(figsize=(10, 5))
anxiety_col = df.columns[3]  
sns.histplot(df[anxiety_col], kde=True, color='blue', bins=10)
plt.title('Anxiety Level Distribution')
plt.xlabel('Anxiety Scale')
plt.ylabel('Count')
plt.grid(True)
plt.show()

# Compute and plot the feature correlation matrix using numerical features
plt.figure(figsize=(10, 8))
numerical_df = df.select_dtypes(include=[np.number])
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
# =====================================================================
# STEP 3: DATA PREPROCESSING, FEATURE SCALING, AND TRAIN-TEST SPLIT
# =====================================================================
# Construct a dynamic binary target classification matrix if not present
if 'Risk_Bracket' not in df.columns:
    df['Risk_Bracket'] = np.where((df.iloc[:, 3] > 5) | (df.iloc[:, 4] == 'Yes'), 1, 0)

# Isolate independent analytical variables from the target output label
X = df.select_dtypes(include=[np.number]).drop('Risk_Bracket', axis=1, errors='ignore')
y = df['Risk_Bracket']

# Normalize continuous attributes into a standardized [0, 1] range
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Partition data into 70% training and 30% testing validation vectors
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)
print(f"X_train shape: {X_train.shape}, X_test shape: {X_test.shape}")
```

```python
# =====================================================================
# STEP 4: TRAIN RANDOM FOREST CLASSIFIER AND EVALUATE DIAGNOSTICS
# =====================================================================
# Initialize and train the nonlinear ensemble model on training matrices
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Execute risk level predictions across the testing evaluation set
y_pred = rf_model.predict(X_test)
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Map predictions onto a confusion matrix to evaluate diagnostic accuracy
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Low Risk', 'High Risk'], yticklabels=['Low Risk', 'High Risk'])
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
# =====================================================================
# STEP 5: SERIALIZE AND EXPORT MODEL OBJECTS TO THE BACKEND PIPELINE
# =====================================================================
# Establish target architecture paths and freeze model state to disk
os.makedirs('../backend/models', exist_ok=True)
model_export_path = '../backend/models/model.pkl'

with open(model_export_path, 'wb') as f:
    pickle.dump(rf_model, f)

print(f"Model saved successfully to: {model_export_path}")
```
